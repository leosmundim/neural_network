# Aplicando Modelo Rede Neural

Documentação: https://www.tensorflow.org/api_docs/python/tf/keras/activations

Referência: https://machinelearningmastery.com/tutorial-first-neural-network-python-keras/

```python
X_train = df_dummie.drop('fuel_economy_combined', axis=1)
y_train = df_dummie['fuel_economy_combined']
```

```python
X_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
model = Sequential()
model.add(Dense(512, input_shape=(26,))) # Camada de entrada com 512 neurônios
model.add(Activation('relu')) # Função de ativação ReLU

model.add(Dense(256)) # Primeira camada escondida com 256 neurônios
model.add(Activation('relu'))

model.add(Dense(128)) # Segunda camada escondida com 128 neurônios
model.add(Activation('relu'))

model.add(Dense(1)) # Camada de saída com 1 neurônio para prever os valores de X
model.add(Activation('relu'))
```

```python
model.compile(loss='MAE', optimizer='adam', metrics=['MAE'])
model.fit(X_train, y_train, batch_size=128, epochs=10)
```

```python
y_hat = pd.DataFrame(model.predict(X_train), columns=['y_hat'])

sns.regplot(x=y_hat['y_hat'], y=y_train, color='orange')
sns.scatterplot(x=y_hat['y_hat'], y=y_train)
```

```python
def indicadores (df, y, y_hat):
    n = len(df)
    SQE = sum((y - y_hat)**2)
    QME = SQE/n
    
    SST = sum((y - y.mean())**2)
    QMT = SST/n
    
    R2 = 1-SQE/SST

    print(f'SQE: {SQE}')
    print(f'QME: {QME}')
    print(f'SST: {SST}')
    print(f'QMT: {QMT}')
    print(f'R2: {R2}')
```

```python
indicadores(df, y_train, y_hat.y_hat)
```
