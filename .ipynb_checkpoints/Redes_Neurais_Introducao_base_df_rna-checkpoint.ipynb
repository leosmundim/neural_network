## Aplicando Modelo

```python
X_train = np.array(df.x).astype('float32')
y_train = np.array(df.y).astype('float32')

#X_train = X_train.reshape(X_train.shape[0],1)

X_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
import tensorflow as tf
```

```python
model = Sequential()
model.add(Dense(512, input_shape=(1,))) # Camada de entrada com 512 neurônios
model.add(Activation('relu')) # Função de ativação ReLU

model.add(Dense(256)) # Primeira camada escondida com 256 neurônios
model.add(Activation('relu'))

model.add(Dense(128)) # Segunda camada escondida com 128 neurônios
model.add(Activation('relu'))

model.add(Dense(1)) # Camada de saída com 1 neurônio para prever os valores de X
model.add(Activation('relu'))
```

```python
model.compile(loss='MAE', optimizer='adam', metrics=['MAE'])
model.fit(df.x, df.y, batch_size=128, epochs=500)
```

```python
X = np.array(df.x).astype('float32')

y_hat = model.predict(df.x)
y_hat
```

## epochs

O parâmetro epochs no método fit do modelo de redes neurais indica o número de vezes que o conjunto de treinamento completo será percorrido durante o processo de treinamento da rede. Em outras palavras, cada época representa uma passagem completa por todos os exemplos do conjunto de treinamento.

Durante o treinamento de uma rede neural, o objetivo é ajustar os pesos da rede de forma que ela seja capaz de fazer previsões precisas em dados não vistos. Esse ajuste é realizado usando algoritmos de otimização, como descida de gradiente (gradient descent), que buscam minimizar uma função de perda (loss function) que mede a diferença entre as previsões do modelo e os rótulos (ou saídas esperadas) dos dados de treinamento.

Quando você especifica o número de epochs durante o treinamento, a rede neural passará várias vezes pelo conjunto de treinamento, atualizando seus pesos a cada passagem, até que o número de epochs especificado seja atingido.

Por exemplo, se você definir epochs=10, a rede neural será treinada por 10 épocas, ou seja, ela percorrerá o conjunto de treinamento 10 vezes, realizando atualizações nos pesos a cada passagem. Cada época é composta por várias iterações, onde cada iteração processa um batch de dados (se você estiver usando treinamento em batch).

A escolha adequada do número de epochs depende do problema específico, do tamanho do conjunto de treinamento e da complexidade da arquitetura da rede neural. Muitas vezes, é necessário encontrar um equilíbrio entre subtreinamento (número insuficiente de epochs) e overfitting (número excessivo de epochs).

Uma abordagem comum é monitorar o desempenho do modelo em um conjunto de validação durante o treinamento e interromper o processo de treinamento quando o desempenho no conjunto de validação parar de melhorar (early stopping). Isso ajuda a evitar o overfitting e economiza tempo de treinamento.

Em resumo, o parâmetro epochs controla a quantidade de vezes que a rede neural será treinada com o conjunto de treinamento durante o processo de treinamento.

## batch_size

O parâmetro `batch_size` no método `fit` do modelo de redes neurais é responsável por determinar o tamanho do lote (batch) de exemplos de treinamento que serão usados em cada etapa de atualização dos pesos durante o processo de treinamento.

Em uma abordagem de treinamento em lote (batch training), o conjunto de treinamento é dividido em lotes menores, e a rede neural é atualizada com base no erro médio calculado para cada lote. Essa abordagem tem várias vantagens:

1. Eficiência computacional: Treinar uma rede neural com muitos dados pode ser intensivo computacionalmente, especialmente se todo o conjunto de treinamento for utilizado em cada passo de atualização dos pesos. Ao usar batches menores, você pode ajustar o modelo mais rapidamente.

2. Uso eficiente da memória: Reduzir o tamanho do batch permite que modelos maiores sejam treinados mesmo em sistemas com recursos limitados de memória.

3. Regularização estocástica: A aleatoriedade introduzida pelo treinamento em batches pode ajudar a evitar overfitting, pois a rede não está sendo atualizada com todos os dados ao mesmo tempo.

O tamanho do `batch_size` define quantos exemplos de treinamento são considerados juntos antes de realizar uma atualização dos pesos do modelo. Por exemplo, se você define `batch_size=32`, o modelo será treinado usando 32 exemplos por vez, e as atualizações dos pesos serão feitas após processar cada lote de 32 exemplos.

Quando o tamanho do batch é igual ao tamanho do conjunto de treinamento, o treinamento é chamado de "treinamento em lote completo" (full batch training). Quando o tamanho do batch é 1, o treinamento é chamado de "treinamento online" (online training). O treinamento em batches intermediários é chamado de "treinamento em mini-lotes" (mini-batch training), e é a abordagem mais comum em redes neurais, pois oferece um bom equilíbrio entre eficiência computacional e atualização dos pesos.

A escolha adequada do tamanho do `batch_size` depende do problema específico, do tamanho do conjunto de treinamento, da capacidade de memória do sistema e da arquitetura da rede neural. Tamanhos de batch maiores podem aproveitar melhor a paralelização em GPUs e TPUs, mas também podem requerer mais memória. Tamanhos de batch menores podem ser úteis em problemas com dados não homogêneos ou para regularização estocástica.

Geralmente, é uma boa prática experimentar diferentes tamanhos de batch para encontrar um equilíbrio entre eficiência e desempenho do modelo.
